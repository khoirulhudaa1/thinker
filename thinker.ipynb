```python
import tkinter as tk

def calculate():
    try:
        result = eval(entry.get())
        label_result.config(text=f"Hasil: {result}")
    except Exception as e:
        label_result.config(text="Error!")

root = tk.Tk()
root.title("Kalkulator Sederhana")
#input
entry = tk.Entry(root, width=20)
entry.pack(pady=5)
#tombol hitung
button = tk.Button(root, text="Hitung", command=calculate)
button.pack(pady=5)
#label hasil
label_result = tk.Label(root, text="Hasil:")
label_result.pack(pady=5)
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

def add_task():
    task = entry_task.get()
    if task:
        listbox_tasks.insert(tk.END, task)
        entry_task.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "Task cannot be empty!")

def delete_task():
    try:
        selected_task = listbox_tasks.curselection()[0]
        listbox_tasks.delete(selected_task)
    except IndexError:
        messagebox.showwarning("Warning", "No task selected!")

root = tk.Tk()
root.title("Todo List")

# Input tugas
entry_task = tk.Entry(root, width=30)
entry_task.pack(pady=5)

# Tombol tambah dan hapus
button_add = tk.Button(root, text="Add Task", command=add_task)
button_add.pack(pady=5)

button_delete = tk.Button(root, text="Delete Task", command=delete_task)
button_delete.pack(pady=5)

# List tugas
listbox_tasks = tk.Listbox(root, width=50, height=10)
listbox_tasks.pack(pady=5)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
def save_note():
    note = text_note.get("1.0", tk.END).strip()
    if note:
        with open("notes.txt", "w") as file:
            file.write(note)
        messagebox.showinfo("Success", "Note saved!")
    else:
        messagebox.showwarning("Warning", "Note is empty!")
def load_note():
    try:
        with open("notes.txt", "r") as file:
            content = file.read()
            text_note.delete("1.0", tk.END)
            text_note.insert("1.0", content)
    except FileNotFoundError:
        messagebox.showwarning("Warning", "No saved notes found!")
root = tk.Tk()
root.title("Note Taking App")
#Text widget untuk catatan
text_note = tk.Text(root, width=50, height=15)
text_note.pack(pady=10)
#Tombol save dan load
button_save = tk.Button(root, text="Save Note", command=save_note)
button_save.pack(pady=5)
button_load = tk.Button(root, text="Load Note", command=load_note)
button_load.pack(pady=5)
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
#data akun
users = {"admin": "12345", "user": "password"}
def login():
    username = entry_username.get()
    password = entry_password.get()
    if username in users and users[username] == password:
        messagebox.showinfo("Login Success", f"Welcome, {username}!")
    else:
        messagebox.showerror("Login Failed", "Invalid username or password!")
        
root = tk.Tk()
root.title("Login App")
#label dan entry untuk username
tk.Label(root, text="Username:").pack(pady=5)
entry_username = tk.Entry(root)
entry_username.pack(pady=5)
#label dan etry untuk password
tk.Label(root, text="Password:").pack(pady=5)
entry_password = tk.Entry(root, show="*")
entry_password.pack(pady=5)
#tombol login
button_login = tk.Button(root, text="Login", command=login)
button_login.pack(pady=10)
root.mainloop()
```
